```python
import pandas as pd
import numpy as np
```

```python
Roles = pd.read_csv('../data/Roles.csv')
Roles['Difficulty'] = Roles['Difficulty'].astype(str).replace("nan","")
Roles.head()
```

```python
InputRoles = ["Artist", "Noble", "Poisoner", "Scarlet Woman", "Spy", "Witch", "Prophet", "Sailor", "Soldier", "Tea Lady", "Drunk", "Balloonist", "Flower Girl", "Investigator", "Fang Gu", "Ojo", "Prisdelo", "Vigormortis" "Ravenkeeper", "Magician", "Mayor", "Monk",  "Moleman", "Ogre", "Recluse", "Sweetheart", "Baron"]
```

```python
WorkingScript = Roles[Roles['Role'].isin(InputRoles)]
WorkingScript
```

```python
def print_glossary(working_script):
    message = ""
    groups = working_script.groupby('Alignment', sort=False)
    for name, group in groups:
        message += "\n"
        message += f"# **{name}**\n"
        for role in group.itertuples():
            message += f"**{role.Role}{role.Difficulty}** {role.Description}\n"
    
        
def print_outline(working_script):
    alignment_check = ""
    for i in working_script.index:
        # Checks if there is a new type
        if alignment_check != working_script['Alignment'][i]:
            print("")
            print(f"# **{working_script['Alignment'][i]}**")
            alignment_check = working_script['Alignment'][i]
        
        # Checks the difficulty
        if type(working_script['Difficulty'][i]) == float:
            working_difficulty = ""
        else:
            working_difficulty = working_script['Difficulty'][i]
            
        # Output
        print(f"**{working_script['Role'][i]}{working_difficulty}**")
        
def save_script(working_script, filename):
    working_script.to_csv(f"../data/{filename}.csv", index=False, header=True)
    
def fetch_script(filename):
    df = pd.read_csv(f"../data/{filename}.csv")
    df['Difficulty'] = df['Difficulty'].astype(str).replace("nan","")
    return df

def make_script(input_roles):
    return Roles[Roles['Role'].isin(input_roles)]
```

```python
print_glossary(WorkingScript)
```


# **Town**
**Artistnan** Privately, during the day, you may return home to ask the storyteller any one yes or no question
**Balloonist*** Nightly, learn the name of a living player of a different type from (demon, minion, outsider, or town) than last night. There might be +1 outsider
**Flower Girlnan** Nightly, learn if the demon voted that day
**Investigatornan** Start knowing 1 of 2 people is a specific minion
**Magician*** The minion is told that either you or the demon is their demon, and the demon is told that either you or the minion is their minion
**Mayornan** If nobody is executed in the final 3, town wins. When you are attacked, another town may die instead
**Monknan** Nightly, choose someone to protect from the demon
**Noblenan** Start with a list of 2 good players and 1 evil, you don't know which is which
**Prophetnan** If a team has not yet won 3 days after your death, you return to life and cannot be killed again by the same method
**Sailor*** You can't die. Nightly, choose a living player, who is told they were visited by the Sailor. You or they are drunk until the start of the next night
**Soldiernan** Cannot be killed by the demon at night
**Tea Ladynan** If both of your living neighbors are good, neither can die

# **Outsider**
**Drunknan** You think you are a specific town role that is not in play, but your ability doesn’t work
**Moleman*** Each night, choose three living players (before anyone dies that night). If one is executed the following day, one of the other two, if alive, will be instead
**Ogre*** On the first night, pick a player. You are their alignment for the rest of the game, but are not told what that alignment is
**Reclusenan** May register as some manner of evil to other role abilities, even after dying
**Sweetheartnan** When you die, a player becomes a drunk for the rest of the game

# **Minion**
**Baronnan** +2 Outsiders
**Poisonernan** Nightly, poison 1 player until the next night begins
**Scarlet Womannan** If the demon dies when there are 5 or more people alive, you become the demon
**Spynan** Nightly, learn all roles. You may appear as good to role abilities
**Witchnan** Nightly, choose a player. If they nominate someone, they die. This ability is disabled the moment 3 or less people remain alive

# **Demon**
**Fang Gunan** Nightly, choose a living player to die. The first time per game an outsider is killed like this, the original Fang Gu dies and the targeted outsider becomes the new Fang Gu. +1 Outsider instead of town
**Ojonan** Nightly, choose a role. If a player with that role exists, they die. If nobody has the role, the storyteller decides who dies
**Prisdelo*** Nightly, choose 3 living players. Each is informed that they have been selected. Each must individually choose whether to live or die. If all choose to live, all die
